```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns

data = pd.read_csv(r"C:\Users\om\OneDrive\Desktop\UM Assignments\Employee Dataset\Employee Dataset.csv")

print(data.head())
```

```python
## Create a histogram of the fixed acidity column
plt.figure(figsize=(11,6))
sns.histplot(data=data,x="healthy_eating",color="red" ,edgecolor="linen",bins=5,alpha=0.5  )
plt.title("Histogram of fixed acidity",fontsize=20)
plt.xlabel("fixed acidity",fontsize=15) 
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
round(data["healthy_eating"].mean(),2)
```

```python
data["healthy_eating"].mode()[0]
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x="healthy_eating",color="red" ,edgecolor="linen",bins=5,alpha=0.5  )
plt.title("Histogram of healthy_eating",fontsize=20)
plt.xlabel("fixed acidity",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.axvline(data["healthy_eating"].mean(), ymin=0, ymax=4000 , color="blue",label="Mean",linestyle="--")
plt.axvline(data["healthy_eating"].median(), ymin=0, ymax=4000 , color="green",label="Median",linestyle="--")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data,x="active_lifestyle",color="green" ,edgecolor="linen",bins=5,alpha=0.5  )  
plt.title("Histogram of active_lifestyle",fontsize=20)
plt.xlabel("volatile acidity",fontsize=15)  
plt.ylabel("Count",fontsize=15)
plt.show()
```

```python
## Plot a distplot of the volatile acidity column
plt.figure(figsize=(11,6))
sns.distplot(data["active_lifestyle"],color="blue")
plt.title("Distplot of volatile acidity",fontsize=20)
plt.xlabel("volatile acidity",fontsize=15)
plt.ylabel("Density",fontsize=15)
plt.show()
```

```python
## calculate skewness and kurtosis of the volatile acidity column
data["active_lifestyle"].skew()
data["active_lifestyle"].kurt()
```

```python
## Create a histogram of volatile acidity column with mean and median       
plt.figure(figsize=(11,6))
sns.histplot(data=data,x="active_lifestyle",color="green" ,edgecolor="linen",bins=5,alpha=0.5  )
plt.title("Histogram of volatile acidity",fontsize=20)
plt.xlabel("volatile acidity",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.axvline(data["active_lifestyle"].mean(), ymin=0, ymax=4000 , color="blue",label="Mean")
plt.axvline(data["active_lifestyle"].median(), ymin=0, ymax=4000 , color="green",label="Median")
plt.legend()
plt.show()
```

```python
## Create a histogram of  citric acid column with mean and median

plt.figure(figsize=(11,6))
sns.histplot(data=data,x="salary",color="orange" ,edgecolor="linen",bins=5,alpha=0.5  )
plt.title("Histogram of salary",fontsize=20)
plt.xlabel("citric acid",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.axvline(data["salary"].mean(), ymin=0, ymax=4000 , color="blue",label="Mean")
plt.axvline(data["salary"].median(), ymin=0, ymax=4000 , color="green",label="Median")
plt.legend()
plt.show()
```

```python
## create a distplot of the citric acid column
plt.figure(figsize=(11,6))
sns.distplot(data["salary"],color="orange")
plt.title("Distplot of citric acid",fontsize=20)
plt.xlabel("citric acid",fontsize=15)
plt.ylabel("Density",fontsize=15)
plt.show()
```
